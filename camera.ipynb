```python
import carla
import time
import cv2
import numpy as np
import CarlaCompatible
client = carla.Client("localhost",2000)
world = client.get_world()
map = world.get_map()
spawn_point = map.get_spawn_points()[0]
Lane_Detectors = CarlaCompatible.LaneDetectorSlidingWindow()
```

```python
vehicle = world.get_actors().filter("vehicle.nissan.*")[0]

camera_bp = world.get_blueprint_library().find('sensor.camera.rgb')
camera_bp.set_attribute('sensor_tick', '0.1')

camera_transform = carla.Transform(
    carla.Location(x=1.5, z=1.8),
    carla.Rotation(pitch=-5)
)
```

```python
print(Lane_Detectors.top_left)
print(Lane_Detectors.Lower_S)
print(Lane_Detectors.Lower_V)
```

```python
print(vehicle.get_transform())
print(camera.get_transform())
```

```python
camera = world.spawn_actor(
    camera_bp, 
    camera_transform, 
    attach_to=vehicle, 
    attachment_type=carla.AttachmentType.Rigid
)
```

```python

def process_image(image):
    try:
        # image.save_to_disk("image.png")
        (Lane_Detectors.process_frame(image))
    except Exception:
        import traceback
        traceback.print_exc()
        raise

```

```python

camera.stop()
```

```python
while True:

    if Lane_Detectors.latest_mask is not None:
        cv2.imshow("mask", Lane_Detectors.latest_mask)

    if Lane_Detectors.latest_result is not None:
        cv2.imshow("result", Lane_Detectors.latest_result)

    if Lane_Detectors.latest_sliding_window is not None:
        cv2.imshow("Sliding window", Lane_Detectors.latest_sliding_window)
    if cv2.waitKey(1) == 27:
        break
```

```python
camera.listen(process_image)
```
